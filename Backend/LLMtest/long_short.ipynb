```python
from datasets import load_dataset

# Load the dataset from Hugging Face Hub
dataset = load_dataset("MartinusChoi/FinPilot")

# Inspect the dataset
print(dataset)
```

```python
question=[dataset['train']['text'][i].split("\n\n### Response: ")[0].split("### Instruction: ")[1] for i in range(len(dataset['train']['text']))]
```

```python
print(question[0])
```

```python
from openai import OpenAI
import os

from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)
answer=[]
def generate_haiku(sentence):
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "developer", "content": "You are a helpful assistant. Speak Korean."},
            {
                "role": "user",
                "content": sentence
            }
        ]
    )
    return completion.choices[0].message.content

for i in range(len(question)):
    output=generate_haiku(question[i])
    answer.append(output)

print(answer)
```

```python
import pandas as pd
df = pd.DataFrame({'question':question, 'answer':answer})
df.head()
```

```python
df.to_csv('finpilotQA.csv', index=False)
```

```python
import pandas as pd

csv_data = pd.read_csv("finpilotQA.csv")
csv_data.head()
```

```python
# drop Unnamed: 0 column
csv_data.drop('Unnamed: 0', axis=1, inplace=True)
csv_data.head()
```

```python
csv_data['question'][0]
```

```python
def generate_prompt(texts, targets):
    output_texts = []
    for text, target in zip(texts["question"], targets["answer"]):
        prompt = f"### Instruction: {text}\n\n### Response: {target}<eos>"
        output_texts.append(prompt)
    return output_texts
```

```python
llmset=[]
llmset = generate_prompt(csv_data,csv_data)
```

```python
llmdf=pd.DataFrame(llmset)
#change column names
llmdf.columns = ['text']
llmdf
```

```python
# to csv
llmdf.to_csv("train_dataset.csv", index=False, encoding="utf-8")
```
